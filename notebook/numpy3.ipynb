```python
import numpy as np
arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
```

```python
# 1.Slice and print every third element starting from index 1.
sliced_arr = arr[1::3]
sliced_arr
```

```python
# 2. Print the last 4 elements using negative indexing.
last_four_elements = arr[-4:]
last_four_elements
```

```python
# 3.Replace the values at even indices with -1.
even_values = arr.copy()
even_values[1::2] = -1
even_values
```

```python
# 4.Use fancy indexing to extract the values at positions [2, 4, 6, 8].
indices = [2, 4, 6, 8]
fancy_values = arr[indices]
fancy_values
```
