```python
import numpy as np
```

```python
# defining list as an numpy array 
list = [1,2,3,4,5] 
array = np.array(list)
array
```

```python
# Or tuple (also dict as well)
tuple = (9,8,5,6,3,1) 
array = np.array(tuple)
array
```

```python
# 1D array
array1 = np.array([1,2,3,4,5])

# 2D array
array2 = np.array([[1,2], [3,4]])

# 3D array
array3 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])

print("1D array:",array1)
print("2D array:",array2)
print("3D array:",array3)
```

```python
# Creating from scratch

# array filled with zeros only
array1 = np.zeros(20)
array1
```

```python
# array filled with ones only
array2 = np.ones(20)
array2
```

```python
# array of numbers using arange
array3 = np.arange(20,50,3) # start,stop, steps
a1 = np.arange(100,252)
# array3
a1
```

```python
# array using linspace
array4 = np.linspace(4,10,5) #start,end, no. of elements 
b1 = np.linspace(40,100)
# array4
b1
```
