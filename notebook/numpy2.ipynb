Task: Perform array slicing

```python
# Perform array slicing
arr = np.random.randint(40,80,8)
print(arr)

arr1 = arr[4:8]
print(arr1)
arr2 = arr[:1:-1]
arr2
```

Create an array of number from 1  to 10 and print only even numbers from that

```python
# Create an array of number from 1  to 10 and print only even numbers from that array
arr = np.arange(1,11)
print(arr)
even_arr = arr[arr % 2 == 0]
print(even_arr)

# odd numbers
odd_arr = arr[arr % 2 != 0]
odd_arr
```
